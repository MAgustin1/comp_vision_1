```python
# Se cargan las imagenes y se guardan en un diccionario

img_1 = cv.imread('images/img_1.png')
img_2 = cv.imread('images/img_2.png')
img_3 = cv.imread('images/img_3.png')
img_4 = cv.imread('images/img_4.png')
img_5 = cv.imread('images/img_5.png')
img_6 = cv.imread('images/img_6.jpg')
img_7 = cv.imread('images/img_7.png')

img_1_rgb= cv.cvtColor(img_1, cv.COLOR_BGR2RGB)
img_1_gray= cv.cvtColor(img_1, cv.COLOR_BGR2GRAY)
img_2_rgb= cv.cvtColor(img_2, cv.COLOR_BGR2RGB)
img_2_gray= cv.cvtColor(img_2, cv.COLOR_BGR2GRAY)
img_3_rgb= cv.cvtColor(img_3, cv.COLOR_BGR2RGB)
img_3_gray= cv.cvtColor(img_3, cv.COLOR_BGR2GRAY)
img_4_rgb= cv.cvtColor(img_4, cv.COLOR_BGR2RGB)
img_4_gray= cv.cvtColor(img_4, cv.COLOR_BGR2GRAY)
img_5_rgb= cv.cvtColor(img_5, cv.COLOR_BGR2RGB)
img_5_gray= cv.cvtColor(img_5, cv.COLOR_BGR2GRAY)
img_6_rgb= cv.cvtColor(img_6, cv.COLOR_BGR2RGB)
img_6_gray= cv.cvtColor(img_6, cv.COLOR_BGR2GRAY)
img_7_rgb= cv.cvtColor(img_7, cv.COLOR_BGR2RGB)
img_7_gray= cv.cvtColor(img_7, cv.COLOR_BGR2GRAY)

img_rgb = {
    "image_1": img_1_rgb, 
    "image_2": img_2_rgb, 
    "image_3": img_3_rgb, 
    "image_4": img_4_rgb, 
    "image_5": img_5_rgb,
    "image_6": img_6_rgb,
    "image_7": img_7_rgb
}
```

```python
# Se carga el template y se distinguen los bordes
template_o = cv.imread('template/pattern.png',0)
template = cv.Canny(template_o, 50, 200)

plt.figure()
plt.imshow(template, cmap='gray')
```

# Parte 1

```python

def edge_detect(img,min,max):

    # Detección de los bordes por método Canny
    img_edge = cv.Canny(img,min,max)

    return img_edge

def scales_pyramid(img, begin, end):
    print(f'Tamaño original: {img.shape}')
    img_to_scale = img.copy()
    scales = np.linspace(begin, end, 5)
    resized_imgs = list()

    # Se genera la pirámide
    print(f'Dimensiones generadas:')
    for scale in scales:
        resized_imgs.append(cv.resize(img_to_scale, None, fx=scale, fy=scale))
        print(resized_imgs[-1].shape)

    return resized_imgs

```

Detección de los logos

```python
w, h = template.shape[::-1]

methods = ['cv.TM_CCOEFF_NORMED']

imgs_rez = {
    "image_1": resized_imgs_1, 
    "image_2": resized_imgs_2, 
    "image_3": resized_imgs_3, 
    "image_4": resized_imgs_4, 
    "image_5": resized_imgs_5,
    "image_6": resized_imgs_6,
    "image_7": resized_imgs_7
}

for name, img in imgs_rez.items():
    for meth in methods:
        img_salida = cv.resize(img_rgb[name], [img.shape[1],img.shape[0]], interpolation = cv.INTER_AREA)
        method = eval(meth)
        
        # Se aplica la coincidencia de patrones
        res = cv.matchTemplate(img, template, method)
        
        # Valores máximos y mínimos
        min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
        
        top_left = max_loc
        
        print("Nivel de confianza: " + str("{:.2f}".format(max_val)))

        # Ubicación de coincidencia
        bottom_right = (top_left[0] + w, top_left[1] + h)


        cv.rectangle(img_salida,top_left, bottom_right, 0, 15)
        
        # Grafica del procesamiento y la salida
        
        # Resultado
        plt.subplot(121),plt.imshow(res,cmap = 'gray')
        plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
        
        # Imagen original con recuadros
        plt.subplot(122),plt.imshow(img_salida)
        plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
        
        img_salida = cv.cvtColor(img_salida,cv.COLOR_BGR2RGB)
        file_name = 'output/simple_match/' + name + '_output.png'
        cv.imwrite(file_name,img_salida)

        plt.suptitle(meth)
        plt.show()
```

# Parte 2

Detección múltiple para una imágen

```python
img_3_copy = resized_imgs_3.copy()
template = template.copy()
output = cv.resize(img_rgb['image_3'], [img_3_copy.shape[1],img_3_copy.shape[0]], interpolation = cv.INTER_AREA)

w, h = template.shape[::-1]
res = cv.matchTemplate(img_3_copy,template,cv.TM_CCOEFF_NORMED)
# Fijamos un umbral de "coincidencia"
threshold = 0.11


loc = np.where( res >= threshold)

for pt in zip(*loc[::-1]):
    cv.rectangle(output, pt, (pt[0] + w, pt[1] + h), (0), 10)

cv.imwrite('multiple_matching.png',output)
plt.imshow(output,cmap = 'gray')
```

# Parte 3

Multiple matching para todas las imágenes

```python
w, h = template.shape[::-1]

methods = ['cv.TM_CCOEFF_NORMED']

imgs_rez = {
    "image_1": resized_imgs_1, 
    "image_2": resized_imgs_2, 
    "image_3": resized_imgs_3, 
    "image_4": resized_imgs_4, 
    "image_5": resized_imgs_5,
    "image_6": resized_imgs_6,
    "image_7": resized_imgs_7
}

for name, img in imgs_rez.items():
    for meth in methods:
        
        img_salida = cv.resize(img_rgb[name], [img.shape[1],img.shape[0]], interpolation = cv.INTER_AREA)
        method = eval(meth)
        
        # Se aplica la coincidencia de patrones
        res = cv.matchTemplate(img, template, method)
        
        # Fijamos un umbral de "coincidencia"
        threshold = 0.11

        
        loc = np.where( res >= threshold)
        for pt in zip(*loc[::-1]):
            cv.rectangle(img_salida, pt, (pt[0] + w, pt[1] + h), (0), 20)

        # Grafica del procesamiento y la salida

        # Resultado
        plt.subplot(121),plt.imshow(res,cmap = 'gray')
        plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
        
        # Imagen original con recuadros
        plt.subplot(122),plt.imshow(img_salida)
        img_salida = cv.cvtColor(img_salida,cv.COLOR_BGR2RGB)
        plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
        
        file_name = 'output/multiple_match/' + name + '_output.png'
        cv.imwrite(file_name,img_salida)
        plt.show()
```
