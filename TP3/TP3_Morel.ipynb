```python
def gaussian_pyramid(img, num_levels):
    lower = img.copy()
    gaussian_pyr = [lower]
    for i in range(num_levels):
        lower = cv.pyrDown(lower)
        gaussian_pyr.append(lower)
    return gaussian_pyr
```

```python
w, h = template.shape[::-1]

methods = ['cv.TM_CCOEFF_NORMED']
imgs_rgb = {
    "image_1": img_1_rgb, 
    "image_2": img_2_rgb, 
    "image_3": img_3_rgb, 
    "image_4": img_4_rgb, 
    "image_5": img_5_rgb,
    "image_6": img_6_rgb,
    "image_7": img_7_rgb
}
imgs_rez = {
    "image_1": resized_imgs_1, 
    "image_2": resized_imgs_2, 
    "image_3": resized_imgs_3, 
    "image_4": resized_imgs_4, 
    "image_5": resized_imgs_5,
    "image_6": resized_imgs_6,
    "image_7": resized_imgs_7
}

for name, img in imgs_rez.items():
    for meth in methods:
        img_salida = cv.resize(imgs_rgb[name], [img.shape[1],img.shape[0]], interpolation = cv.INTER_AREA)
        method = eval(meth)
        
        # Aplicamos la coincidencia de patrones
        #--------------------------------------
        res = cv.matchTemplate(img, template, method)
        
        # Encontramos los valores máximos y mínimos
        min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
        
        top_left = max_loc
        
        print("Nivel de confianza: " + str("{:.2f}".format(max_val)))

        # Marcamos el lugar donde lo haya encontrado
        #----------------------------------------
        bottom_right = (top_left[0] + w, top_left[1] + h)


        cv.rectangle(img_salida,top_left, bottom_right, 0, 15)
        
        # Graficamos el procesamiento y la salida
        #----------------------------------------
        
        # Resultado de coincidencia
        plt.subplot(121),plt.imshow(res,cmap = 'gray')
        plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
        
        # Imagen original con recuadros
        plt.subplot(122),plt.imshow(img_salida)
        plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
        
        img_salida = cv.cvtColor(img_salida,cv.COLOR_BGR2RGB)
        file_name = 'output/simple_match/' + name + '_output.png'
        cv.imwrite(file_name,img_salida)

        plt.suptitle(meth)
        plt.show()
```

```python
img_ej2 = cv.imread('images/coca_multi.png')
img_ej2_rgb= cv.cvtColor(img_ej2, cv.COLOR_BGR2RGB)
img_ej2_gray= cv.cvtColor(img_ej2, cv.COLOR_BGR2GRAY)
template_ej2 = cv.imread('template/pattern.png',0)
```

```python
def scales_pyramid(img, begin, end):
    print(f'Tamaño original: {img.shape}')
    img_to_scale = img.copy()
    scales = np.linspace(begin, end, 5)
    resized_imgs = list()
    print(f'Dimensiones generadas:')
    
    for scale in scales:
        resized_imgs.append(cv.resize(img_to_scale, None, fx=scale, fy=scale))
        print(resized_imgs[-1].shape)

    return resized_imgs
```

```python
img_aux = img_ej2_gray.copy()

gray = (int) (np.average(template_ej2[90:100,50:70].flatten()))
img_aux[img_ej2_gray > 180] = gray
img_aux[img_ej2_gray < 180] = 180

print("Template shape: " + str(template.shape))
pyramid = scales_pyramid(img_aux, 3,4)
size = len(pyramid)
img_ej2_res = pyramid[3]
```

```python
img_gray = img_ej2_res.copy()
template = template.copy()
output = cv.resize(img_ej2_rgb, [img_gray.shape[1],img_gray.shape[0]], interpolation = cv.INTER_AREA)

w, h = template.shape[::-1]
res = cv.matchTemplate(img_gray,template,cv.TM_CCOEFF_NORMED)
# Fijamos un umbral de "coincidencia"
threshold = 0.33


loc = np.where( res >= threshold)

for pt in zip(*loc[::-1]):
    cv.rectangle(output, pt, (pt[0] + w, pt[1] + h), (0), 10)

cv.imwrite('multiple_matching.png',output)
plt.imshow(output,cmap = 'gray')
```
